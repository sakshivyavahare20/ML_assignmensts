```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
heart = pd.read_csv(r"C:\Users\SRINIDHI\OneDrive\Desktop\ML\MLAssignments\Assignment 2\Heart (1).csv")
heart.head()
```

```python
heart.isnull().sum()
```

```python
heart.describe()
```

# Bar Plot 

```python


# Bar Chart: Distribution of Chest Pain Types
plt.figure(figsize=(8, 6))
heart['ChestPain'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Distribution of Chest Pain Types', fontsize=16)
plt.xlabel('Chest Pain Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Histogram 

```python
plt.figure(figsize=(8, 6))
plt.hist(heart['Age'], bins=10, color='green', edgecolor='black', alpha=0.7)

# Adding labels and title
plt.title('Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# line chart 

```python
age_maxhr = heart.groupby('Age')['MaxHR'].mean()

plt.figure(figsize=(10, 6))
plt.plot(age_maxhr.index, age_maxhr.values, marker='o', color='blue')

# Adding labels and title
plt.title('Average Max Heart Rate by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Max Heart Rate (bpm)', fontsize=14)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# pie chart

```python

ahd_counts = heart['AHD'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(ahd_counts, labels=ahd_counts.index, autopct='%1.1f%%', startangle=140, colors=['orange', 'skyblue'])
plt.title('Proportion of AHD Categories', fontsize=16)
plt.tight_layout()
plt.show()
```

# Box Plot

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Chol', data=heart, palette='coolwarm')

# Adding labels and title
plt.title('Cholesterol Levels by AHD', fontsize=16)
plt.xlabel('Heart Disease (AHD)', fontsize=14)
plt.ylabel('Cholesterol (mg/dL)', fontsize=14)
plt.tight_layout()
plt.show()
```

# scatter plot

```python
# Scatter Plot: Cholesterol vs MaxHR
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Chol', y='MaxHR', hue='AHD', data=heart, palette='coolwarm', s=100)

# Adding labels and title
plt.title('Cholesterol vs Max Heart Rate', fontsize=16)
plt.xlabel('Cholesterol (mg/dL)', fontsize=14)
plt.ylabel('Max Heart Rate (bpm)', fontsize=14)
plt.legend(title='AHD')
plt.tight_layout()
plt.show()
```
